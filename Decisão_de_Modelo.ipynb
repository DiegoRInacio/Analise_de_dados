__Exercício 2:__ Realizar um círculo branco no gráfico do conteúdo em anexo, indicando os pontos $(x1, x2)$ tais que $x1^2$ + $x2^2$ $\sim$ 10. Ou seja, pares $(x1, x2)$ cujo valor da f é próximo de 10.

Considere que "próximo" neste caso significa que o valor resultante da f não pode ser maior, em módulo do que 1. Isto é $-1<f(x) - 10 < 1$, ou seja,  $9 <f(x)<11$

***

###### Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Funções usadas

```python
def f(x1, x2):
    return x1**2 + x2**2

def create_meshgrid(x1_lims, x2_lims):
    x1mesh, x2mesh = np.meshgrid(x1_lims, x2_lims)
    return np.c_[x1mesh.ravel(), x2mesh.ravel()]
```

##### Definindo tamanho dos vetores para obter uma malha contínua:

```python
x1_lims = np.arange(-10, 10, 0.02)
x2_lims = np.arange(-10, 10, 0.02)

print('O tamanho dos vetores, x1 = {} e x2 = {}'.format(len(x1_lims), len(x2_lims)))
```

##### Criando uma malha usando os vetores definidos acima:

```python
mesh = create_meshgrid(x1_lims, x2_lims)

print('Apresentando dimensões do meshgrid: {}'.format(mesh.shape))
```

##### Aplicando a função em todos os pontos da malha e salvando os valores em lista

```python
L = []

x1_w = [0]
x2_w = [1]

for v in mesh:
    x1 = v[0]
    x2 = v[1]
    value_f = f(x1, x2)
    if 9 < value_f < 11:
        x1_w.append(x1)
        x2_w.append(x2)
        
    L.append(value_f)
```

```python
plt.figure(figsize=[15, 10])
plt.scatter(mesh[:, 0], mesh[:, 1], c = L, cmap = plt.cm.rainbow)
plt.scatter(x1_w, x2_w, c = 'white')
plt.colorbar()
plt.show()
```
