<h3>Tratamento adequado para os dados faltantes é:</h3>
<br>
    <h4>Foi criada uma função de preprocessamento para tratamento das variáveis, sendo que a função irá traser o seguinte resultado:</h4>
        <h4>1. Idade: inserir idade média já que são 99 dados faltante de 29170.</h4>
        <h4>2. Nome: deletar o atributo nome uma vez que ele esta por inteiro nulo</h4>
        <h4>3. Dívida: deletar o atributo dívida pois, assim como nome, ele esta por inteiro nulo</h4>

```python
#Função de préprocessamento/tratamento

def preprocessamento(df, cols_num, cols_cat):
    dff = df.copy()
    temp = (df.isnull().sum() / df.shape[0]) >= 0.8
    var_80PorcFalt = temp.loc[temp == True].index.tolist()
    for v in var_80PorcFalt:
        dff = dff.drop(v, axis = 1)
    
    cols_num = list(set(cols_num).intersection(set(dff.columns.tolist())))
    cols_cat = list(set(cols_cat).intersection(set(dff.columns.tolist())))
    
    for v in cols_num:
        dff[v] = dff[v].fillna(dff[v].mean())
    
    for v in cols_cat:
        mode_v = dff[v].value_counts().idxmax()
        dff[v] = dff[v].fillna(mode_v)
    
    return dff
```

```python
df1 = preprocessamento(df = df,
                      cols_num = ['idade', 'tempo_educacao', 'horas_por_semana'],
                      cols_cat = ['estado_civil', 'cor', 'sexo', 'salario_anual', 'nome', 'dívida'])

df1
```

<h3>9 Detectando e eliminando os <strong>Outliers</strong> da variável 'Idade'</h3>

```python
iqr = np.percentile(df.idade, 75) - np.percentile(df.idade, 25)

iqr
```

```python
#calculando os limite dos intervalo interquatil

def detecta_outlier(x, limite = 1.5):
    iqr = np.percentile(x, 75) - np.percentile(x, 25)
    limite_inf = np.maximum(np.percentile(x, 25) - limite * iqr, np.min(x))
    limite_sup = np.minimum(np.percentile(x, 75) + limite * iqr, np.max(x))
    return np.where((x < limite_inf) | (x > limite_sup), 1, 0)
```

```python
df['outlier_idade'] = detecta_outlier(df.idade)
```
